# **1. Công thức toán liên quan**


> ## **1.1. Chuyển đổi ảnh màu sang ảnh xám**
*   Công thức: $Gray = 0.2989 \times R + 0.5870 \times G + 0.1140 \times B$
*   Trong đó R, G, B là giá trị của các pixel ảnh màu, và Gray là giá trị pixel ảnh xám tương ứng.


> ## **1.2. Kết hợp các channel màu để tạo ảnh màu từ ảnh xám**
*   Công thức: $RGB=(Gray,Gray,Gray)$

# **2. Hiện thực và đưa ra kết quả**

```python
# Thư viện
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
# Đọc file ảnh màu
img = cv2.imread('/image.jpg')

# Hiển thị ảnh màu
cv2_imshow(img)
```

```python
# Đọc file ảnh xám
gray_img = cv2.imread('/image.jpg',0)

# Hiển thị ảnh xám
cv2_imshow(gray_img)
```

> ## **2.1. Chuyển đổi ảnh màu sang ảnh xám**

```python
# Lấy 3 ảnh xám từ 3 kênh màu b,g,r của ảnh màu
r,g,b = img[:,:,0], img[:,:,1], img[:,:,2]

# Hiển thị ảnh xám của 3 kênh màu
cv2_imshow(r)
cv2_imshow(g)
cv2_imshow(b)
```


> > ### *2.1.1. Sử dụng công thức toán*

```python
# Tạo ảnh xám theo công thức toán
gray_1 = 0.2989*r + 0.5870*g + 0.1140*b

# Hiển thị ảnh xám
cv2_imshow(gray_1)
```

### *2.1.2. Sử dụng built-in function*

```python
# Tạo ảnh xám bằng built-in function của OpenCV
gray_2 = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# Hiển thị ảnh xám
cv2_imshow(gray_2)
```

> ## **2.2. Kết hợp các channel màu để tạo ảnh màu từ các ảnh xám**

> > ### *2.2.1. Sử dụng built-in function*

```python
img_combined_1 = cv2.merge([r,g,b])

# Hiển thị ảnh màu kết hợp
cv2_imshow(img_combined_1)
```

> > ### *2.2.2. Sử dụng công thức*

```python
# Lấy chiều cao, chiều rộng và chiều sâu của ảnh xám
h, w = gray_img.shape

# Tạo một ảnh màu mới có kích thước và kiểu dữ liệu tương tự ảnh xám
img_combined_2 = np.zeros((h, w, 3), dtype=np.uint8)

# Hiển thị ảnh mới
cv2_imshow(img_combined_2)
```

```python
# Gán giá trị màu cho từng kênh màu
img_combined_2[:,:,0] = r  # Kênh màu xanh lam
img_combined_2[:,:,1] = g  # Kênh màu xanh lá cây
img_combined_2[:,:,2] = b  # Kênh màu đỏ

# Hiển thị ảnh kết hợp
cv2_imshow(img_combined_2)
```
